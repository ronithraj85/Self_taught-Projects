```python
import pandas as pd
```

```python
train_data = pd.read_excel('Data_Train.xlsx')
train_data.head()
```

```python
test_data = pd.read_excel('Test_set.xlsx')
test_data.head(2)
```

```python
train_data.describe()
```

```python
train_data['Journey_Day'] = train_data.Date_of_Journey.str.split('/').str[0]
train_data['Journey_Month'] = train_data.Date_of_Journey.str.split('/').str[1]
train_data['Journey_Year'] = train_data.Date_of_Journey.str.split('/').str[2]
```

```python
test_data['Journey_Day'] = test_data.Date_of_Journey.str.split('/').str[0]
test_data['Journey_Month'] = test_data.Date_of_Journey.str.split('/').str[1]
test_data['Journey_Year'] = test_data.Date_of_Journey.str.split('/').str[2]
```

```python
train_data.drop(columns='Date_of_Journey',axis=1,inplace=True)
test_data.drop(columns='Date_of_Journey',axis=1,inplace=True)
```

```python
train_data.Journey_Day = train_data.Journey_Day.astype('int')
train_data.Journey_Month = train_data.Journey_Month.astype('int')
train_data.Journey_Year = train_data.Journey_Year.astype('int')

test_data.Journey_Day = test_data.Journey_Day.astype('int')
test_data.Journey_Month = test_data.Journey_Month.astype('int')
test_data.Journey_Year = test_data.Journey_Year.astype('int')
```

```python
train_data.Arrival_Time = train_data.Arrival_Time.str.split(' ').str[0]
test_data.Arrival_Time = test_data.Arrival_Time.str.split(' ').str[0]
```

```python
train_data[train_data.Total_Stops.isnull()]
```

```python
train_data.Total_Stops = train_data.Total_Stops.fillna('1 stop')
```

```python
train_data[train_data.Total_Stops.isnull()]
```

```python
train_data.Total_Stops = train_data.Total_Stops.replace('non-stop','0 stop')

test_data.Total_Stops = test_data.Total_Stops.replace('non-stop','0 stop')
```

```python
train_data['Arrival_Hours'] = train_data.Arrival_Time.str.split(':').str[0]
train_data['Arrival_Minutes'] = train_data.Arrival_Time.str.split(':').str[1]
train_data.Arrival_Hours = train_data.Arrival_Hours.astype(int)
train_data.Arrival_Minutes = train_data.Arrival_Minutes.astype(int)
train_data.drop(['Arrival_Time'],axis=1,inplace=True)


test_data['Arrival_Hours'] = test_data.Arrival_Time.str.split(':').str[0]
test_data['Arrival_Minutes'] = test_data.Arrival_Time.str.split(':').str[1]
test_data.Arrival_Hours = test_data.Arrival_Hours.astype(int)
test_data.Arrival_Minutes = test_data.Arrival_Minutes.astype(int)
test_data.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
train_data['Departure_Hours'] = train_data.Dep_Time.str.split(':').str[0]
train_data['Departure_Minutes'] = train_data.Dep_Time.str.split(':').str[1]
train_data.Departure_Hours = train_data.Departure_Hours.astype(int)
train_data.Departure_Minutes = train_data.Departure_Minutes.astype(int)
train_data.drop(['Dep_Time'],axis=1,inplace=True)

test_data['Departure_Hours'] = test_data.Dep_Time.str.split(':').str[0]
test_data['Departure_Minutes'] = test_data.Dep_Time.str.split(':').str[1]
test_data.Departure_Hours = test_data.Departure_Hours.astype(int)
test_data.Departure_Minutes = test_data.Departure_Minutes.astype(int)
test_data.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
train_data['Route_1'] = train_data.Route.str.split('→ ').str[0]
train_data['Route_2'] = train_data.Route.str.split('→ ').str[1]
train_data['Route_3'] = train_data.Route.str.split('→ ').str[2]
train_data['Route_4'] = train_data.Route.str.split('→ ').str[3]
train_data['Route_5'] = train_data.Route.str.split('→ ').str[4]
train_data.drop(['Route'],axis=1, inplace=True)

test_data['Route_1'] = test_data.Route.str.split('→ ').str[0]
test_data['Route_2'] = test_data.Route.str.split('→ ').str[1]
test_data['Route_3'] = test_data.Route.str.split('→ ').str[2]
test_data['Route_4'] = test_data.Route.str.split('→ ').str[3]
test_data['Route_5'] = test_data.Route.str.split('→ ').str[4]
test_data.drop(['Route'],axis=1, inplace=True)

```

```python
train_data.Route_1.fillna('None',inplace=True)
train_data.Route_2.fillna('None',inplace=True)
train_data.Route_3.fillna('None',inplace=True)
train_data.Route_4.fillna('None',inplace=True)
train_data.Route_5.fillna('None',inplace=True)

test_data.Route_1.fillna('None',inplace=True)
test_data.Route_2.fillna('None',inplace=True)
test_data.Route_3.fillna('None',inplace=True)
test_data.Route_4.fillna('None',inplace=True)
test_data.Route_5.fillna('None',inplace=True)
```

```python
train_data['Duration_hours'] = train_data.Duration.str.split('h').str[0]
test_data['Duration_hours'] = test_data.Duration.str.split('h').str[0]
```

```python
train_data['Duration_minutes'] = train_data.Duration_hours.str.split('m').str[0]
train_data.Duration_hours.replace("5m","0",inplace=True)
train_data.Duration_minutes.replace("","0",inplace=True)
train_data.Duration_hours = train_data.Duration_hours.astype(int)
train_data.Duration_minutes = train_data.Departure_Minutes.astype(int)
train_data.drop(['Duration'],axis=1,inplace=True)
```

```python
test_data['Duration_minutes'] = test_data.Duration_hours.str.split('m').str[0]
test_data.Duration_hours.replace("5m","0",inplace=True)
test_data.Duration_minutes.replace("","0",inplace=True)
test_data.Duration_hours = test_data.Duration_hours.astype(int)
test_data.Duration_minutes = test_data.Departure_Minutes.astype(int)
test_data.drop(['Duration'],axis=1,inplace=True)
```

```python
train_data.drop(['Additional_Info','Route_1','Route_2','Route_3','Route_4','Route_5'],axis=1,inplace=True)
test_data.drop(['Additional_Info','Route_1','Route_2','Route_3','Route_4','Route_5'],axis=1,inplace=True)
```

```python
train_data.drop(['Journey_Year'], axis=1, inplace=True)
# test_data.drop(['Journey_Year'], axis=1, inplace=True)
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
train_data.Total_Stops = label_encoder.fit_transform(train_data.Total_Stops)
test_data.Total_Stops = label_encoder.transform(test_data.Total_Stops)
```

```python
train_data.head()
```

```python
train_Airline = train_data[["Airline"]]

train_Airline = pd.get_dummies(train_Airline, drop_first=True)

train_Airline.head()

test_Airline = test_data[["Airline"]]

test_Airline = pd.get_dummies(test_Airline, drop_first=True)

test_Airline.head()
```

```python
train_Source = train_data[["Source"]]

train_Source = pd.get_dummies(train_Source)

train_Source.head()

test_Source = test_data[["Source"]]

test_Source = pd.get_dummies(test_Source)

test_Source.head()
```

```python

train_Destination = train_data[["Destination"]]

train_Destination = pd.get_dummies(train_Destination)

train_Destination.head()

test_Destination = test_data[["Destination"]]

test_Destination = pd.get_dummies(test_Destination)
test_Destination.head()
```

```python
train_data.drop(['Airline','Source','Destination'],axis=1,inplace=True)
train_data = pd.concat([train_data, train_Airline, train_Source, train_Destination],axis=1)

test_data.drop(['Airline','Source','Destination'],axis=1,inplace=True)
test_data = pd.concat([test_data, test_Airline, test_Source, test_Destination],axis=1)
```

```python
train_data.drop(['Destination_New Delhi'], axis=1,inplace=True)

test_data.drop(['Destination_New Delhi'], axis=1,inplace=True)
```

```python
X = train_data.drop(['Price'],axis=1)
Y = train_data.Price
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor()
random_forest_reg.fit(X,Y)
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
```

```python

#No of trees
n_estimators = [int(x) for x in np.linspace(start=100, stop=1300, num=13)]

#Number of features to consider at every split
max_features = ['auto','sqrt','log2']

#Maximum no.of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num=7)]

#minimum no.of samples required to split a node
min_samples_split = [2,5,10,15,100]

#minimum no.of samples required at each leaf node
min_samples_leaf = [1,2,5,10]

random_grid = {
                'n_estimators' : n_estimators,
                 'max_features' : max_features,
                 'max_depth':max_depth,
                 'min_samples_leaf' : min_samples_leaf,
                 'min_samples_split' : min_samples_split
                }
```

```python

random_rf = RandomizedSearchCV(RandomForestRegressor(), param_distributions=random_grid, scoring='neg_mean_squared_error',n_iter=10,cv=5, verbose= 2 ,random_state=12)
```

```python
random_rf.fit(X,Y)
```

```python
predictions = random_rf.predict(test_data)
```
